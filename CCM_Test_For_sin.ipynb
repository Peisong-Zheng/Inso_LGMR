```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a sine signal
n_points = 200
t = np.linspace(0, 4 * np.pi, n_points)  # 0 to 4
x = np.sin(t)

# 2. Prepare figure and subplots
fig, axes = plt.subplots(2, 3, figsize=(10, 6))
axes = axes.flatten()

taus = [1, 2, 3, 4, 5, 6]

for i, tau in enumerate(taus):
    # For E=2, the time-delay embedding is simply:
    #   X-axis: x(t)
    #   Y-axis: x(t + )
    # We lose 'tau' data points at the end when building pairs

    tau=tau*1

    X = x[:-tau]    # from index 0 to (end - tau)
    Y = x[tau:]     # from index tau to end

    ax = axes[i]
    ax.plot(X, Y, marker="o", linestyle="-")
    ax.set_xlabel(r"$x(t)$")
    ax.set_ylabel(rf"$x(t + \tau)$")
    ax.set_title(rf"$\tau = {tau}$")

plt.tight_layout()
plt.show()
```
