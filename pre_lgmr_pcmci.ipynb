```python
import pandas as pd
import numpy as np
from scipy.signal import hilbert
import matplotlib.pyplot as plt


# data_array has [X, Y]
data_array = df[['X', 'Y']].values
print("\nInitial data_array (columns: [X, Y]):")
print(data_array)

# ---------------------------------------------------------------------------
# STEP 1: For the second column (Y), calculate diff and store it back.
# Because diff is one element shorter, we drop a row (here, we drop the last row).
# ---------------------------------------------------------------------------
y_diff = np.diff(data_array[:, 1])  # diff of column 1 (Y)
# data_array is currently shape (N, 2); y_diff is shape (N-1,)
data_array = data_array[:-1, :]     # drop the last row from data_array
data_array[:, 1] = y_diff           # replace Y with diff(Y)
print("\nAfter diff on the 2nd column and dropping last row:")
print(data_array)

# ---------------------------------------------------------------------------
# STEP 2: For the first column in the original DataFrame (X),
#         perform a Hilbert transform, extract amplitude & phase,
#         and store them as two new columns in data_array,
#         while deleting the original first column (X) from data_array.
# ---------------------------------------------------------------------------
# Perform the Hilbert transform on the original X (has length N).
x_original = df['X'].values
analytic_signal = hilbert(x_original)
amplitude = np.abs(analytic_signal)
phase = np.angle(analytic_signal)

# Because we already dropped 1 row in data_array, we do the same for amplitude & phase
amplitude = amplitude[:-1]
phase = phase[:-1]

# Now remove the old first column (X) from data_array.
# Currently data_array = [X, diff(Y)], shape (N-1, 2).
# We only keep the second column (diff(Y)) for now:
diff_y = data_array[:, 1].reshape(-1, 1)

# Insert amplitude and phase as two new columns before diff(Y).
data_array = np.column_stack((amplitude, phase, diff_y))

print("\nFinal data_array (columns: [amplitude, phase, diff(Y)]):")
print(data_array)

# ---------------------------------------------------------------------------
# Plot amplitude and phase
# ---------------------------------------------------------------------------
plt.figure()
plt.plot(amplitude)
plt.title('Amplitude')
plt.show()

plt.figure()
plt.plot(phase)
plt.title('Phase')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tigramite
from tigramite import data_processing as pp
from tigramite import plotting as tp
from tigramite.pcmci import PCMCI
from tigramite.independence_tests import parcorr, gpdc, cmiknn
import pandas as pd

# Create a Tigramite DataFrame object
var_names = [f"Box {k+1}" for k in range(2)]

# create adataframe called 'df4stationary_check' with the data_array and the variable names
# df4stationary_check= pd.DataFrame(data_array, columns=var_names)

dataframe_TEMP = pp.DataFrame(data_array, datatime=np.arange(data_array.shape[0]), var_names=var_names)
tp.plot_timeseries(dataframe_TEMP,label_fontsize=15,tick_label_size=10)





ParCorr = parcorr.ParCorr(significance='analytic')
pcmci = PCMCI(
    dataframe=dataframe_TEMP,
    cond_ind_test=ParCorr,
    verbosity=1)


matrix_lags = None #np.argmax(np.abs(correlations), axis=2)
tp.plot_scatterplots(dataframe=dataframe_TEMP, add_scatterplot_args={'matrix_lags':matrix_lags}); plt.show()
correlations = pcmci.get_lagged_dependencies(tau_max=20, val_only=True)['val_matrix']
lag_func_matrix = tp.plot_lagfuncs(val_matrix=correlations, setup_args={'var_names':var_names, 
                                    'x_base':5, 'y_base':.5}); plt.show()

pcmci.verbosity = 1
tau_min=1
tau_max=10
results = pcmci.run_pcmci(tau_min=tau_min, tau_max=tau_max, pc_alpha=None)
# q_matrix = pcmci.get_corrected_pvalues(p_matrix=results['p_matrix'], tau_max=tau_max, fdr_method='fdr_bh')

# create a figure
fig, ax = plt.subplots(1, 1, figsize=(20, 20), dpi=100)
tp.plot_graph(
    fig_ax=(fig, ax),
    graph = results['graph'],
    val_matrix=results['val_matrix'],
    var_names=var_names,
    link_colorbar_label='MCI',
    node_colorbar_label='auto-MCI',
    link_label_fontsize=20,
    label_fontsize=20,
    tick_label_size=14,
    node_label_size=10,
    edge_ticks=0.5,
    node_ticks=0.5,
    node_size=0.2
    )
```
